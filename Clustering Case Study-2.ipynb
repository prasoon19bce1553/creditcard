## Define features

```python
x= df.drop(columns="CUST_ID")
```

## Standardise data

```python
from sklearn.preprocessing import StandardScaler
stdScalar= StandardScaler()
```

```python
stdScalar.fit(x)
x_std= stdScalar.transform(x)
```

## Build cluster model

```python
from sklearn.cluster import KMeans
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x_std)
    distortions.append(kmeanModel.inertia_)
```

```python
distortions
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

## We can observe that the “elbow” is the number 4 which is best or optimal for this case. Now we can run a K-Means using as n_clusters the number 4.

```python
kmeanModel = KMeans(n_clusters=4)
```

```python
kmeanModel.fit(x_std)
```

```python
y_k= kmeanModel.predict(x_std)
```

```python
y_k
```

## Visualising Data based on cluster
